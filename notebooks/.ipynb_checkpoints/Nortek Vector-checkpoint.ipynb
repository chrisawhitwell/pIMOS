## Important note on Dolfyn

`dat.u`, `dat.v` and `dat.w` are just shortcuts for `dat.vel[0, :]`, `dat.vel[1, :]` and `dat.vel[2, :]`

check `dat['props']['coord_sys']` to know what's what.

# Now try some other rotations

```python

datcc = datc.copy()
try:
    beam2inst(datcc, reverse=True)
    print('Success!!')
except:
    print('Failed in converting from INST to BEAM because we are not in the right coordinate system.')
    
try:
    inst2earth(datcc)
    print('Success!!')
except:
    print('Failed in converting from INST to EARTH because we are not in the right coordinate system.')
    
try:
    inst2earth(datcc, reverse=True)
    print('Success!!')
except:
    print('Failed in converting from EARTH to INST because we are not in the right coordinate system.')
    
try:
    beam2inst(datcc)
    print('Success in converting BEAM to INST!!')
    inst2earth(datcc)
    print('Success in converting INST to EARTH!!')
except:
    print('Failed in converting from BEAT to INST to EARTH.')
    
print('And we are now in {} coordinates'.format(datcc['props']['coord_sys']))
```

```python
%matplotlib inline
plt.figure(figsize=(16,3))
plt.plot(time, datcc.u[:], label='dolfyn E?')
plt.plot(time, datcc.v[:], label='dolfyn N?')
plt.plot(time, datcc.w[:], label='dolfyn U?')
plt.grid()
```
