# Example notebook for processing SBE data with pIMOS

Included instruments:
- SBE56
- SBE39 T
- SBE39 TP
- SBE37 CTD
- And of course our very special friend, SBE37 4567 CT


Notebook demonstrates:
- API for reading
- All variables/netcdf files CF compliant [compliant enough]
- API for QAQC [note there are no automated codes for this, this just shows how such a code should interface with the xarray objects]
- API for exporting 

At the bottom ALL seabird instruments are read and processed in a loop.

```python
# import seabird as sbd
# import crude_readers.read.SEABIRD_37_39_56 as read_sbd

import pIMOS.xrwrap.SEABIRD_37_39_56 as wrap_sbd
import os
import matplotlib.pyplot as plt
import datetime
import numpy as np 

os.path.exists(r'./KISSME samples/SBE39/3724.cnv')

# O Drive is //drive.irds.uwa.edu.au/CEME-BBL-001
pIMOS_export_dir = r'O:\Data\pIMOS\KISSME'

def pIMOS_export(rr, model, serial):
    
    folder = os.path.join(pIMOS_export_dir, model)
    if not os.path.exists(folder):
        os.mkdir(folder)

    print(folder)

    rr.folder = folder
    rr.file_ = serial
    rr.export()
        
```

# QAQC
Apply QAQC code to everything prior to Apil 2 at 5:00 and replot.

Note the qc_Temperature and qc_Pressure fields in the dataset above. These control the QAQC. Use the get_qaqc_var method to update these. Update both using '*'.

```python
rr.update_qc_flag('*', 
                  'time',
                  np.datetime64(datetime.datetime(2017, 1, 1)),
                  np.datetime64(datetime.datetime(2017, 4, 2, 5)),
                 1)

# Now use the 'get_qaqc_var' method to pull the QAQC'd value
rr.get_qaqc_var('Temperature').plot(figsize=(15, 3))
plt.grid()
plt.xlim(rr.ds.time[[0, -1]])
plt.show()

rr.ds['Temperature'].plot(figsize=(15, 3), label='QAQC')
# You can still pull the raw data like this
rr.get_qaqc_var('Temperature').plot(label='Raw')
plt.grid()
plt.xlim(rr.ds.time[[0, -1]])
plt.show()
```

```python
os.path.exists(csvfile)

filetypes = ['SBE39 TP', 'SBE39 T', 'SBE39-ext T', 'SBE37 CT'];

instr_files = []
with open(csvfile) as f:
    for line in f.readlines():
        line = line.split(',')
        if line[9] in ['SBE39 TP', 'SBE39 T', 'SBE39-ext T']:
            print(line)
            
            asc_filename = os.path.join(root, line[8], line[7])
            print(asc_filename)
            
            sn = line[10]

            if os.path.exists(asc_filename):
                print('File exists')
            else:
                raise(Exception)
                
#             mcout_asc = read_sbd.parse_seabird_asc(asc_filename)
            
            folder, file = os.path.split(asc_filename)
            rr = wrap_sbd.SEABIRD_37_39_56(folder, file, model=line[9][0:5])
            
            pIMOS_export(rr, line[9][0:5], sn)
            
```
