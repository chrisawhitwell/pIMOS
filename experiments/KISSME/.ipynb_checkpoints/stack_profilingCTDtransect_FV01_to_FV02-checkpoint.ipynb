```python
def get_files(archive_dir, transect, instruments):
    files = []
    for instrument in instruments:
        instr_dir = os.path.join(archive_dir, instrument) 
        wc = instr_dir+'/*KISSME2017*[[]{}[]]*.nc'.format(transect)

        files += glob.glob(wc)
    
    if len(files) == 0:
        raise(Exception("No files found"))
        
    return files

stack_variables = ['Fluorescence', 'Temperature', 'Temperature2', 'Salinity', 'Beam_attenuation']
```

```python
files = get_files(archive_dir_fv01, 'SP250_12hr', ['Solander CTD'])
rr_SP250_12hr      = st.from_fv01_archive_slow(files, stack_variables, np.arange(5, 250, 1))

files = get_files(archive_dir_fv01, 'SP250_12hr_neap', ['Solander CTD'])
rr_SP250_12hr_neap = st.from_fv01_archive_slow(files, stack_variables, np.arange(5, 250))

files = get_files(archive_dir_fv01, 'UWAB_Transect', ['Solander CTD'])
rr_UWAB_Transect   = st.from_fv01_archive_slow(files, stack_variables, np.arange(5, 500, 5))

files = get_files(archive_dir_fv01, 'UWAD_Transect', ['Solander CTD'])
rr_UWAD_Transect   = st.from_fv01_archive_slow(files, stack_variables, np.arange(5, 500, 5))
```

```python
def transect_plot(rr, zmax=250, plotby='time', funky=False):
    
    if plotby.lower() == 'time':
        xnew = rr.ds['time']
        xlab = ''
    elif plotby.lower() == 'dist':
        xnew = rr.ds['dist_approx']
        xlab = 'Approx distance from beginning of transect'
    else:
        error
        
    if funky:
        app = '_i'
        print('FUNKY')
    else:
        app = ''
        
    plt.figure(figsize=(20, 3))
    plt.pcolormesh(xnew, rr.ds['z'+app], rr.ds['Temperature'+app], cmap='Spectral_r', shading='gouraud')
    plt.colorbar(label='Temp [deg C]')
    plt.ylabel('Depth [m]')
    plt.ylim(zmax, 0)

    plt.figure(figsize=(20, 3))
    plt.pcolormesh(xnew, rr.ds['z'+app], rr.ds['Temperature2'+app], cmap='Spectral_r', shading='gouraud')
    plt.colorbar(label='Temp [deg C]')
    plt.ylabel('Depth [m]')
    plt.ylim(zmax, 0)

    plt.figure(figsize=(20, 3))
    plt.pcolormesh(xnew, rr.ds['z'+app], rr.ds['Fluorescence'+app], cmap='Greens', shading='gouraud')
    plt.colorbar(label='Fl [mg m$^{-3}$]')
    plt.ylabel('Depth [m]')
    plt.ylim(zmax, 0)

    plt.figure(figsize=(20, 3))
    plt.pcolormesh(xnew, rr.ds['z'+app], rr.ds['Beam_attenuation'+app], cmap='magma', shading='gouraud')
    plt.colorbar(label='Att [m$^{-1}$]')
    plt.ylabel('Depth [m]')
    plt.xlabel(xlab)
    plt.ylim(zmax, 0)
    
```
