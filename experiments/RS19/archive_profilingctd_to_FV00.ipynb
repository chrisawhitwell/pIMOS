```python
db_data['ctd_log']

direc = r'D:\Experiments\kissme17\ProfileData\CTD'
direc = r'G:\Experiments\kissme17\ProfileData\CTD'
instrument = 'Solander CTD'

for index, row in db_data['ctd_log'].iterrows():

#     instrument = row['InstrumentType']
#     sn = row['SerialNo']
    filename = row['Filename']
    total_depth = row['Total Depth Sounder depth plus 2m Xduser offset (m)']
#     attributes = {
#         'site': row['Site'],
#         'site_station': row['StationID'],
#         'nominal_latitude': float(row['Latitude']),
#         'nominal_longitude': float(row['Longitude']),
#         'nominal_site_depth': float(row['Depth']),
#         'nominal_instrument_height_asb': float(row['InstrumentHeight']),
#         'nominal_instrument_orientation': row['Orientation'],
#         'timezone': row['TimeZone'],
#         'instrument_model': row['InstrumentType'],
#         'instrument_serial_number': row['SerialNo'],
#         }

    attributes = {
        'project': 'KISSME2017',
        'raw_file_name': filename,
        'is_profile_data': 1,
        'timezone': 'UTC',
        'nominal_site_depth': total_depth}
    
    cnvfile = os.path.join(direc, filename+'.cnv')
    
    rr, ds = solander_ctd.from_cnvfile(cnvfile)
    rr.update_attributes_with_dict(attributes)

    comment_fields = ['UWA Niskin Bottle No. + Depth (m)', 'UWA Bottles', 'LISST Raw Filename', 'Comment']
    for field in comment_fields:
        print()
        if type(row[field]) == str:
            comment = field + ': ' + row[field]
        elif np.isnan(row[field]):
            comment = field + ': - '
        else:
            comment = field + ': ' + str(row[field])

        rr.add_comment('UWA', comment)

    rr.ds['t190C'].plot()
    plt.grid()
    plt.show()
    
    fn = filename
    ai.pIMOS_export(rr, archive_dir, instrument, fn)
    
```
